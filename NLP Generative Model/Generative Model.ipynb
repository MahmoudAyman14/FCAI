```python
import wikipedia

def generate_document(topic):
    try:
        # Fetch Wikipedia content
        page = wikipedia.page(topic)
        title = page.title
        content = page.content

        # Concatenate title and content
        document = f"Title: {title}\n\n{content}"

        return document

    except wikipedia.exceptions.PageError:
        print("Error: Page not found. Please check the topic.")

    except wikipedia.exceptions.DisambiguationError as e:
        print(f"Error: DisambiguationError - {e.options}. Please specify a more precise topic.")

# Example usage
topic = "Deep Learning"
document_1 = generate_document(topic)
print(document_1)
```

```python
import wikipedia

def generate_document(topic):
    try:
        # Fetch Wikipedia content
        page = wikipedia.page(topic)
        title = page.title
        content = page.content

        # Concatenate title and content
        document = f"Title: {title}\n\n{content}"

        return document

    except wikipedia.exceptions.PageError:
        print("Error: Page not found. Please check the topic.")

    except wikipedia.exceptions.DisambiguationError as e:
        print(f"Error: DisambiguationError - {e.options}. Please specify a more precise topic.")

# Example usage
topic = "artifial intelligence"
document_4 = generate_document(topic)
print(document_4)

```

```python
import re
import nltk
nltk.download('wordnet')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import numpy as np
nltk.download('punkt')

nltk.download('stopwords')

```

```python
def process_text(text):
    text = re.sub(r'\s+', ' ', text, flags=re.I) # Remove extra white space from text

    text = re.sub(r"https?://\S+|www\.\S+"," ",text)

    # Removing html tags
    text = re.sub(r"<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});"," ",text)

    text = re.sub(r'\W', ' ', str(text)) # Remove all the special characters from text

    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text) # Remove all single characters from text
    word = re.escape("title")
    pattern = r'\b' + word + r'\b'
    text = re.sub(pattern, '', text, flags=re.IGNORECASE)
    text = re.sub(r'[^a-zA-Z\s]', '', text) # Remove any character that isn't alphabetical
    pattern1 = r'\s'
    # Use re.sub() to replace the first whitespace with an empty string
    text = re.sub(pattern1, '', text, count=1)
    word2 = re.escape("Deep learning")
    pattern2 = r'\b' + word2 + r'\b'
    text = re.sub(pattern2, '', text, flags=re.IGNORECASE,count=1)
    text = re.sub(pattern1, '', text, count=1)

    text = text.lower()

    # words = word_tokenize(text)

    # lemmatizer = WordNetLemmatizer()
    # words = [lemmatizer.lemmatize(word) for word in words]

    # stop_words = set(stopwords.words("english"))
    # Words = [word for word in words if word not in stop_words]

    # Words = [word for word in Words if len(word) > 3]

    # indices = np.unique(Words, return_index=True)[1]
    cleaned_text = text

    return cleaned_text
```

```python
tokenize_text = [process_text(text) for text in documents]
```

```python
print(tokenize_text)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, GRU, LSTM, Dense, Embedding,Input
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(tokenize_text)
total_words = len(tokenizer.word_index) + 1

# Convert text to sequences
sequences = tokenizer.texts_to_sequences(tokenize_text)
```

```python
print(tokenizer.word_index)
```

```python
maxlen = 20  # Choose the length of input sequences
input_sequences = []
labels = []
for sequence in sequences:
    for i in range(maxlen, len(sequence)):
        # Create input sequence and corresponding label
        input_seq = sequence[i - maxlen:i]
        label = sequence[i]
        input_sequences.append(input_seq)
        labels.append(label)
```

```python
# max_sequence_length = max(len(seq) for seq in input_sequences)
input_sequences = np.array(input_sequences)
labels = np.array(labels)

# Pad input sequences
input_sequences = pad_sequences(input_sequences, maxlen=maxlen, padding='pre')
```

```python
# print(max_sequence_length)
# print(len(input_sequences))
```

```python
labels = tf.keras.utils.to_categorical(labels, num_classes=total_words)
```

```python
model_rnn = Sequential()
model_rnn.add(Embedding(input_dim=total_words, output_dim=300, input_length=maxlen))
model_rnn.add(SimpleRNN(500, return_sequences=False))
model_rnn.add(Dense(total_words, activation='softmax'))
model_rnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print(model_rnn.summary())
```

```python
model_rnn.fit(input_sequences, labels, epochs=15)
```

```python
# Generate text using the trained models
def generate_text(seed_text, model, max_sequence_len, num_words):
    for _ in range(num_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_sequence_len, padding='pre')
        predicted = np.argmax(model.predict(token_list), axis=-1)
        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break
        seed_text += " " + output_word
    return seed_text

# Example of generating text with each model
generated_text_rnn = generate_text("Ai", model_rnn, maxlen, num_words=20)


print("Generated Text (SimpleRNN):", generated_text_rnn)
```

```python
# Assuming 'df' is your DataFrame containing the 'epoch' column
import pandas as pd

# Grouping by 'epoch' and creating a dictionary of DataFrames for each unique epoch
Events = {epoch: group for epoch, group in df.groupby('epoch')}

# Accessing individual DataFrames for each distinct epoch
for epoch_value, epoch_df in Events.items():
    print(f"DataFrame for Epoch {epoch_value}:")
    print(epoch_df)
    print("\n")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dense, Dropout
from sklearn.utils.class_weight import compute_class_weight

# Define Morlet wavelet function
def morlet_wavelet(x, w=6.0):
    return np.cos(w * x) * np.exp(-0.5 * (x ** 2))

# Define Morlet wavelet transform
def morlet_wavelet_transform(signal, scales):
    result = np.zeros((len(scales), len(signal)))

    for i, scale in enumerate(scales):
        # Scale the Morlet wavelet
        scaled_wavelet = morlet_wavelet(np.linspace(-5, 5, len(signal)) / scale)
        scaled_wavelet /= np.linalg.norm(scaled_wavelet)  # Normalize the wavelet

        # Convolve the signal with the scaled wavelet
        result[i, :] = np.convolve(signal, scaled_wavelet, mode='same')

    return result


# Step 1: Prepare Data
X_list, y_list = [], []

for epoch_value, epoch_df in Events.items():
    X_epoch = epoch_df.drop(columns=['patient', 'label', 'epoch'])
    y_epoch = epoch_df['label'].map(label_mapping)

    # Apply Wavelet Transform
    scales = np.arange(1, 11)  # Example scales, you can adjust them
    X_wavelet = np.apply_along_axis(lambda x: morlet_wavelet_transform(x, scales), axis=1, arr=X_epoch.values)

    X_list.append(X_wavelet)
    y_list.append(y_epoch.values)

X_all = np.concatenate(X_list, axis=0)
y_all = np.concatenate(y_list, axis=0)

# Reshape to have a consistent input shape for the neural network
X_all = X_all.reshape(X_all.shape[0], -1)

# Step 2: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size=0.2, random_state=42)

# Calculate class weights
class_counts = np.bincount(y_train)
total_samples = len(y_train)
class_weights = total_samples / (len(np.unique(y_train)) * class_counts)
class_weight_dict = dict(zip(np.unique(y_train), class_weights))

model = Sequential()
model.add(Dense(512, input_dim=X_train.shape[1], activation='relu'))  
model.add(Dense(256, activation='relu')) 
model.add(Dropout(0.25))
model.add(Dense(128, activation='relu'))  
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(32, activation='relu'))
model.add(Dense(4, activation='softmax')) 


model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, class_weight=class_weight_dict)

# Step 4: Evaluate the Model
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)  

# Compute evaluation metrics
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(classification_rep)
```

## Read  html page from url

```python
import requests

# URL of the webpage
url = 'https://en.wikipedia.org/wiki/Main_Page'

# Make a request to the webpage
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract HTML content
    html_content = response.text
    # Print or do something with the HTML content
    print(html_content)
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
import re
```

```python
#Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

#Extract TEXT from HTML page e.g. (p or headings)


```python
def get_html_text(url):
    # Fetch HTML content from URL
    response = requests.get(url)
    return response.text
# URL of the HTML page
url = "https://en.wikipedia.org/wiki/Main_Page"

# Get HTML content from URL
html_text = get_html_text(url)

# Parse HTML using BeautifulSoup
soup = BeautifulSoup(html_text, 'html.parser')

# Extract text from HTML
#page_text = soup.get_text()
# Extract text from specific HTML elements (e.g., paragraphs or headings)
text_elements = soup.find_all(['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6'])

# Combine text from all selected elements
page_text = ' '.join([element.get_text() for element in text_elements])
print(page_text)
```

```python
def clean_text(text):
    # Remove symbols and characters that are not alphabetic
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
    return cleaned_text
# Clean text
cleaned_text = clean_text(page_text)
print("Cleaned Text:", cleaned_text)
```

#Normalization: make all the data to lower case


```python
def normalize_text(text):
    # Convert text to lowercase
    return text.lower()
# Normalize text
normalized_text = normalize_text(cleaned_text)
print("Normalized Text:", normalized_text)
```

# Tokenization: split the data to words

```python
def tokenize_text(text):
    # Tokenize text into words
    return word_tokenize(text)
# Tokenize text
tokenized_text = tokenize_text(normalized_text)
print("Tokenized Text:", tokenized_text)
```

# Lemmatization or Stemming: return each word to origin.

```python
def lemmatize_text(words):
    # Initialize lemmatizer
    lemmatizer = WordNetLemmatizer()
    # Lemmatize each word
    return [lemmatizer.lemmatize(word) for word in words]
lemmatized_text = lemmatize_text(tokenized_text)
print("Lemmatized Text:", lemmatized_text)
```

# Stop words: remove stop words from the data

```python
def remove_stopwords(words):
    # Get English stopwords
    stop_words = set(stopwords.words('english'))
    # Remove stopwords from words
    return [word for word in words if word not in stop_words]
# Remove stopwords
cleaned_text_no_stopwords = remove_stopwords(lemmatized_text)
print("Text without Stopwords:", cleaned_text_no_stopwords)
```

## Unique Word: the output of the assignment to get all the unique words on the data.

```python
# Extract unique words
unique_words = set(cleaned_text_no_stopwords)
print("Unique Words:", unique_words)
```

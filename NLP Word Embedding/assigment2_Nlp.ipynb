```python
import re
def clean_text(text):
    # Remove symbols and characters that are not alphabetic
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
    return cleaned_text
# Clean text
cleaned_text = clean_text(document_1)
cleaned_text_2 = clean_text(document_2)
cleaned_text_3 = clean_text(document_3)
print("Cleaned Text:", cleaned_text)
print("Cleaned Text:", cleaned_text_2)
print("Cleaned Text:", cleaned_text_3)



```

```python
def normalize_text(text):
    # Convert text to lowercase
    return text.lower()
# Normalize text
normalized_text = normalize_text(cleaned_text)
normalized_text_2 = normalize_text(cleaned_text_2)
normalized_text_3= normalize_text(cleaned_text_3)
print("Normalized Text:", normalized_text)
print("Normalized Text:", normalized_text_2)
print("Normalized Text:", normalized_text_3)
```

```python
def tokenize_text(text):
    # Tokenize text into words
    return word_tokenize(text)
# Tokenize text
tokenized_text = tokenize_text(normalized_text)
tokenized_text_2 = tokenize_text(normalized_text_2)
tokenized_text_3 = tokenize_text(normalized_text_3)

print("Tokenized Text:", tokenized_text)
print("Tokenized Text:", tokenized_text_2)
print("Tokenized Text:", tokenized_text_3)

```

```python
import spacy

# Load the spaCy English model
nlp = spacy.load('en_core_web_sm')

# Lemmatize each word in the list
lemmatized_tokens = []
lemmatized_tokens_2 = []
lemmatized_tokens_3= []


for word in tokenized_text:
    doc = nlp(word)
    lemmatized_tokens.append(doc[0].lemma_ if doc else '')
for word in tokenized_text_2:
    doc = nlp(word)
    lemmatized_tokens_2.append(doc[0].lemma_ if doc else '')
for word in tokenized_text_3:
    doc = nlp(word)
    lemmatized_tokens_3.append(doc[0].lemma_ if doc else '')

# Print the original and lemmatized text
print("Original Words:", tokenized_text)
print("Lemmatized Words:", lemmatized_tokens)

print("Original Words:", tokenized_text_2)
print("Lemmatized Words:", lemmatized_tokens_2)

print("Original Words:", tokenized_text_3)
print("Lemmatized Words:", lemmatized_tokens_3)

```

```python
def remove_stopwords(words):
    # Get English stopwords
    stop_words = set(stopwords.words('english'))
    # Remove stopwords from words
    return [word for word in words if word not in stop_words]
# Remove stopwords
cleaned_text_no_stopwords = remove_stopwords(lemmatized_tokens)
print("Text without Stopwords:", cleaned_text_no_stopwords)
cleaned_text_no_stopwords_2 = remove_stopwords(lemmatized_tokens_2)
print("Text without Stopwords:", cleaned_text_no_stopwords_2)
cleaned_text_no_stopwords_3 = remove_stopwords(lemmatized_tokens_3)
print("Text without Stopwords:", cleaned_text_no_stopwords_3)

```

```python
def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict
```

```python
def computeIDF(documents):
    import math
    N = len(documents)

    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1

    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict
```

```python
idfs = computeIDF([numOfWordsA, numOfWordsB,numOfWordsc])
idfs
```

```python
def computeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    for word, val in tfBagOfWords.items():
        tfidf[word] = val * idfs[word]
    return tfidf
```
